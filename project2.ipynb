Inner join of tables store and staff where from staff_ids are the same. Only printed manager_id and first name where first name was "Mike"  

```python
!psql -c "SELECT pagila.store.manager_staff_id, pagila.staff.first_name pagila FROM pagila.store JOIN pagila.staff ON pagila.store.store_id = pagila.staff.store_id WHERE pagila.staff.first_name = 'Mike' ORDER BY pagila.staff.manager_staff_iod"   
```

Inner join of table of city and country where the country ids are the same. Prints city and country where country is Mexico, ordered by city name 

```python
!psql -c "SELECT pagila.city.city, pagila.country.country FROM pagila.country JOIN pagila.city ON pagila.country.country_id = pagila.city.country_id WHERE pagila.country.country = 'Mexico' ORDER BY pagila.city.city" 
```

Inner join of film and actor tables when they have same film id and we print title and actor id where the actor id is greater than 5, Ordered by film title. (limited to 50 results) 

```python
!psql -c "SELECT pagila.film.title, pagila.film_actor.actor_id FROM pagila.film_actor JOIN pagila.film ON pagila.film_actor.film_id = pagila.film.film_id WHERE pagila.film_actor.actor_id >= 5 ORDER BY pagila.film.title LIMIT 50" 
```

Right of customer and address table through their mutual address id. prints any customer id over 550 and order by first name 

```python
!psql -c "SELECT pagila.customer.first_name, pagila.customer.last_name, pagila.address.postal_code, pagila.customer.customer_id FROM pagila.customer RIGHT JOIN pagila.address ON pagila.address.address_id = pagila.customer.address_id  WHERE pagila.customer.customer_id > 550 ORDER BY pagila.customer.first_name"
```

Left join of store and customer prints store id, manager_staff_id and customer_id. Where pagila store_id is greater than zero and orders by store id. 

```python
!psql -c "SELECT pagila.store.store_id, pagila.store.manager_staff_id, pagila.customer.customer_id FROM pagila.store LEFT JOIN pagila.customer ON pagila.store.address_id = pagila.customer.address_id WHERE pagila.store.manager_staff_id > 0 ORDER BY pagila.store.store_id"
```

Full join of language or film table where language ids matches and film id is greater than 400 ordered by name 

```python
!psql -c "SELECT pagila.language.name, pagila.film.title FROM pagila.language FULL JOIN pagila.film ON pagila.language.language_id = pagila.film.language_id WHERE pagila.film.film_id > 400 ORDER BY pagila.film.film_id LIMIT 10"
```

self join of film table to see if the films have the same rating and not the same film id. shows title and rating. (limit of 10) 

```python
!psql -c "SELECT p1.title, p1.rating  FROM pagila.film p1 JOIN pagila.film p2 ON p1.film_id != p2.film_id AND p1.rating = p2.rating LIMIT 10"
```
